##`Q-1:` Rectangle Class
1. Write a Rectangle class in Python language, allowing you to build a rectangle with length and width attributes.

2. Create a Perimeter() method to calculate the perimeter of the rectangle and a Area() method to calculate the area of ​​the rectangle.

3. Create a method display() that display the length, width, perimeter and area of an object created using an instantiation on rectangle class.

Eg.
After making above classes and methods, on executing below code:-
```
my_rectangle = Rectangle(3 , 4)
my_rectangle.display()
```

`Output:`
```
The length of rectangle is:  3
The width of rectangle is:  4
The perimeter of rectangle is:  14
The area of rectangle is:  12
```


````python
class Rectangle:
  def __init__(self, length, width):
    self.length = length
    self.width = width

  def parimeter(self):
    return 2*(self.length + self.width)

  def area(self):
    return self.length * self.width

  def display(self):
    print(f"The length of rectangle is: {self.length}")
    print(f"The width of rectangle is: {self.width}")
    print(f"The perimeter of rectangle is: {self.parimeter()}")
    print(f"The area of rectangle is: {self.area()}")

rect = Rectangle(3, 4)
rect.display()
````

##`Q-2: Bank Class`

1. Create a Python class called `BankAccount` which represents a bank account, having as attributes: `accountNumber` (numeric type), `name` (name of the account owner as string type), `balance`.
2. Create a constructor with parameters: `accountNumber, name, balance`.
3. Create a `Deposit()` method which manages the deposit actions.
4. Create a `Withdrawal()` method  which manages withdrawals actions.
5. Create an `bankFees()` method to apply the bank fees with a percentage of 5% of the balance account.
6. Create a `display()` method to display account details.
Give the complete code for the  BankAccount class.

Eg.
After making above classes and methods, on executing below code:-
```
newAccount = BankAccount(2178514584, "Mandy" , 2800)

newAccount.Withdrawal(700)

newAccount.Deposit(1000)

newAccount.display()
```

`Output:`
```
Account Number :  2178514584
Account Name :  Mandy
Account Balance :  3100 ₹
```

````python
# write your code here
class BankAccount:
  def __init__(self, accountNumber, name, balance) -> None:
    self.accountNumber = accountNumber
    self.name = name
    self.balance = balance

  def withdrawal(self, amount):
    self.balance -= amount


  def deposit(self, amount):
    self.balance += amount

  def bankFees(self):
    self.balance -= 0.05 * self.balance

  def display(self):
    print(f"Account Number: {self.accountNumber} \nAccount Name: {self.name} \nAccount Balance: {self.balance}")

  def __str__(self) -> str:
    return f"Account Number: {self.accountNumber} \nAccount Name: {self.name} \nAccount Balance: {self.balance}"

newAccount = BankAccount(2178514584, "Mandy" , 2800)

newAccount.withdrawal(700)

newAccount.deposit(1000)

newAccount.display()
````

````python
# claude
# stand alone class, number not as a member


import math


class Computation:
    """Computation class - performs various calculations on integers."""

    # 1. Default constructor (no parameters)
    def __init__(self):
        pass

    # 2. Factorial of n
    def factorial(self, n):
        if n < 0:
            raise ValueError("n must be a non-negative integer")
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result

    # 3. Sum of first n natural numbers: 1 + 2 + ... + n
    def naturalSum(self, n):
        total = 0
        for i in range(1, n + 1):
            total += i
        return total
        # Alternatively: return n * (n + 1) // 2

    # 4. Test if n is prime
    def testPrime(self, n):
        if n < 2:
            return False
        for i in range(2, int(math.sqrt(n)) + 1):
            if n % i == 0:
                return False
        return True

    # Helper: greatest common divisor (Euclid's algorithm)
    def _gcd(self, a, b):
        a, b = abs(a), abs(b)
        while b != 0:
            a, b = b, a % b
        return a

    # 5. Test whether two numbers are prime to each other (coprime)
    def testPrims(self, a, b):
        return self._gcd(a, b) == 1

    # 6a. Display the multiplication table of a given integer
    def tableMult(self, n):
        print(f"Multiplication table of {n}:")
        for i in range(1, 11):
            print(f"{n} x {i} = {n * i}")

    # 6b. Display all multiplication tables from 1 to 9
    def allTablesMult(self):
        for n in range(1, 10):
            self.tableMult(n)
            print()

    # 7a. Static method: get all divisors of n
    @staticmethod
    def listDiv(n):
        Ldiv = []
        for i in range(1, n + 1):
            if n % i == 0:
                Ldiv.append(i)
        return Ldiv

    # Helper for static prime check (used by listDivPrim)
    @staticmethod
    def _isPrimeStatic(n):
        if n < 2:
            return False
        for i in range(2, int(math.sqrt(n)) + 1):
            if n % i == 0:
                return False
        return True

    # 7b. Static method: get all prime divisors of n
    @staticmethod
    def listDivPrim(n):
        primeDivisors = []
        for d in Computation.listDiv(n):
            if Computation._isPrimeStatic(d):
                primeDivisors.append(d)
        return primeDivisors


# Demo / test driver
if __name__ == "__main__":
    c = Computation()

    print("Factorial of 5 =", c.factorial(5))
    print("Sum of first 10 naturals =", c.naturalSum(10))
    print("Is 17 prime?", c.testPrime(17))
    print("Is 20 prime?", c.testPrime(20))
    print("Are 4 and 9 prime to each other?", c.testPrims(4, 9))
    print("Are 6 and 9 prime to each other?", c.testPrims(6, 9))
    print()

    c.tableMult(7)
    print()

    c.allTablesMult()

    print("Divisors of 36 =", Computation.listDiv(36))
    print("Prime divisors of 36 =", Computation.listDivPrim(36))
````

##`Q-4`: Build flashcard using class in Python.

Build a flashcard using class in python. A flashcard is a card having information on both sides, which can be used as an aid in memoization. Flashcards usually have a question on one side and an answer on the other.

**Example 1:**

Approach:

- Create a class named FlashCard.
- Initialize dictionary fruits using __init__() method. Here you have to define fruit name as key and it's color as value. E.g., {"Banana": "yellow", "Strawberries": "pink"}
- Now randomly choose a pair from fruits by using _random_ module and store the key in variable _fruit_ and _value_ in variable color.
- Now prompt the user to answer the color of the randomly chosen fruit.
- If correct print correct else print wrong.

Output:
```bash
welcome to fruit quiz
What is the color of Strawberries
pink
Correct answer
Enter 0, if you want to play again: 0
What is the color of watermelon
green
Correct answer
Enter 0, if you want to play again: 1
```

````python
# write your code here
class FlashCard:
  def __init__(self) -> None:
    self.fruits = {"Banana": "yellow", "Strawberries": "pink", "Watermelon": "green", "Mango": "yellow", "Pineapple": "green", "Apple": "red", "Orange": "orange", "Grapes": "green", "Kiwi": "brown", "Pear": "green"}
    print("welcome to fruit quiz")

  def play(self):
    import random
    while True:
      fruit, color = random.choice(list(self.fruits.items()))
      answer = input(f"What is the color of {fruit}? ")
      if answer == color:
        print("Correct answer")
      else:
        print("Wrong answer")
      option = int(input("Enter 0, if you want to play again: 0"))
      if option != 0:
        break

flashcard = FlashCard()
flashcard.play()
````

## `Q-5:` Problem 5 based on OOP Python.

TechWorld, a technology training center, wants to allocate courses for instructors. An instructor is identified by name, technology skills, experience and average feedback. An instructor is allocated a course, if he/she satisfies the below two conditions:
- eligibility criteria:
    - if experience is more than 3 years, average feedback should be 4.5 or more
    - if experience is 3 years or less, average feedback should be 4 or more
- he/she should posses the technology skill for the course

Identify the class name and attributes to represent instructors. Write a Python program to implement the class chosen with its attributes and methods.

**Note:**
- Consider all instance variables to be private and methods to be public.
- An instructor may have multiple technology skills, so consider instance variable, technology_skill to be a list.
- *check_eligibility()*: Return true if eligibility criteria is satisfied by the instructor. Else, return false
- *allocate_course(technology)*: Return true if the course which requires the given technology can be allocated to the instructor. Else, return false.

Represent a few objects of the class, initialize instance variables using setter methods, invoke
appropriate methods and test your program.

````python
# write your code here
class Instructor:
  def __init__(self, name, experience, feedback, technology_skills) -> None:
    self.name = name
    self.experience = experience
    self.feedback = feedback
    self.technology_skills = technology_skills.split()

  def check_eligibility(self, course):
    if self.experience > 3 and self.feedback >= 4.5 and course in self.technology_skills:
      return True
    elif self.experience <= 3 and self.feedback >= 4 and course in self.technology_skills:
      return True
    else:
      return False

ahmad = Instructor("ahmad", 4, 4.6, "python javascript")
ahmad.check_eligibility("python")
````

````python
# claude
# a full class with member functions and methods
class Instructor:
    """
    Represents an instructor at TechWorld.

    Attributes (all private):
        __name             : str   - instructor's name
        __technology_skill : list  - list of technology skills the instructor possesses
        __experience       : float - years of experience
        __avg_feedback     : float - average feedback score
    """

    # Default constructor
    def __init__(self):
        self.__name = ""
        self.__technology_skill = []
        self.__experience = 0
        self.__avg_feedback = 0.0

    # ---------- Setters ----------
    def set_name(self, name):
        self.__name = name

    def set_technology_skill(self, technology_skill):
        self.__technology_skill = technology_skill

    def set_experience(self, experience):
        self.__experience = experience

    def set_avg_feedback(self, avg_feedback):
        self.__avg_feedback = avg_feedback

    # ---------- Getters ----------
    def get_name(self):
        return self.__name

    def get_technology_skill(self):
        return self.__technology_skill

    def get_experience(self):
        return self.__experience

    def get_avg_feedback(self):
        return self.__avg_feedback

    # ---------- Business logic methods ----------
    def check_eligibility(self):
        """
        Return True if the instructor meets the eligibility criteria:
        - experience > 3 years  -> avg_feedback must be >= 4.5
        - experience <= 3 years -> avg_feedback must be >= 4.0
        """
        if self.__experience > 3:
            return self.__avg_feedback >= 4.5
        else:
            return self.__avg_feedback >= 4.0

    def allocate_course(self, technology):
        """
        Return True if the instructor:
        - satisfies check_eligibility(), AND
        - possesses the given technology skill
        """
        if self.check_eligibility() and technology in self.__technology_skill:
            return True
        return False

    def __str__(self):
        return (f"Instructor: {self.__name}, "
                f"Skills: {self.__technology_skill}, "
                f"Experience: {self.__experience} years, "
                f"Avg Feedback: {self.__avg_feedback}")


# ---------------- Test / demo driver ----------------
if __name__ == "__main__":

    # Instructor 1: experience > 3, feedback >= 4.5 -> eligible
    instructor1 = Instructor()
    instructor1.set_name("Alice Khan")
    instructor1.set_technology_skill(["Python", "Machine Learning", "SQL"])
    instructor1.set_experience(5)
    instructor1.set_avg_feedback(4.7)

    # Instructor 2: experience > 3, feedback < 4.5 -> NOT eligible
    instructor2 = Instructor()
    instructor2.set_name("Bilal Ahmed")
    instructor2.set_technology_skill(["Java", "Spring Boot"])
    instructor2.set_experience(6)
    instructor2.set_avg_feedback(4.2)

    # Instructor 3: experience <= 3, feedback >= 4.0 -> eligible
    instructor3 = Instructor()
    instructor3.set_name("Sara Malik")
    instructor3.set_technology_skill(["Python", "Django", "React"])
    instructor3.set_experience(2)
    instructor3.set_avg_feedback(4.1)

    # Instructor 4: experience <= 3, feedback < 4.0 -> NOT eligible
    instructor4 = Instructor()
    instructor4.set_name("Usman Tariq")
    instructor4.set_technology_skill(["C++", "Python"])
    instructor4.set_experience(1)
    instructor4.set_avg_feedback(3.8)

    instructors = [instructor1, instructor2, instructor3, instructor4]

    for instructor in instructors:
        print(instructor)
        print("Eligible?      ", instructor.check_eligibility())
        print("Allocate Python?", instructor.allocate_course("Python"))
        print("Allocate Java?  ", instructor.allocate_course("Java"))
        print("-" * 60)
````
