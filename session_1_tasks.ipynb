### Q2:- Write a program that will convert celsius value to fahrenheit.

````python
# Write your code here
celsius = float(input("Enter temperature in celsius: "))
fahrenheit = (celsius * 9/5) + 32
print('%.2f Celsius is: %0.2f Fahrenheit' %(celsius, fahrenheit))
````

### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
# Write your code here
print("Enter first coordinates: ")
x1 = int(input("x1: "))
y1 = int(input("y1: "))
print("Enter second coordinates: ")
x2 = int(input("x2: "))
y2 = int(input("y2: "))

distance = ((x2 - x1)**2 + (y2 - y1)**2)**0.5
print("Distance between two coordinates: ",distance)
````

### Q5:- Write a program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.


````python
# Write your code here
p = float(input("Enter the principle amount: "))
r = float(input("Enter the rate of interest: "))
t = float(input("Enter the time period: "))

si = (p * r * t) / 100
print("Simple Interest: ",si)
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# Write your code here
heads = int(input("Enter number of heads: "))
legs = int(input("Enter number of legs: "))

dogs = (legs - (heads * 2)) // 2
chickens = heads - dogs

print("Dogs: ",dogs)
print("Chickens: ",chickens)
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
# Write your code here
n = int(input("Enter n: "))
sum_square = (n*(n+1)*(2*n+1))/6
print("Sum of squares of first n natural numbers: ",sum_square)
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
# Write your code here
a = int(input("Enter first term: "))
b = int(input("Enter second term: "))
n = int(input("Enter n: "))

d = b - a
an = a + (n-1)*d
print("Nth term of the series: ",an)
````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
# Write your code here
n1 = int(input("Enter numerator of first fraction: "))
d1 = int(input("Enter denominator of first fraction: "))
n2 = int(input("Enter numerator of second fraction: "))
d2 = int(input("Enter denominator of second fraction: "))

num = n1*d2 + n2*d1
den = d1*d2

from fractions import Fraction

# Create a fraction (Numerator: 3, Denominator: 6)
f = Fraction(num, den)

print("Sum of two fractions: ",f)
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
# Write your code here
h = int(input("Enter tank height: "))
l = int(input("Enter tank length: "))
b = int(input("Enter tank breadth: "))

glass_h = int(input("Enter glass height: "))
glass_r = int(input("Enter glass radius: "))

import math
volume_tank = h*l*b
volume_glass = 3.1416 * (glass_r ** 2) * glass_h

ans = volume_tank // volume_glass

print("Tank Volume = ", volume_tank)
print("Glass Volume = ", volume_glass)
print("Number of glasses of milk that can be obtained = ", ans)
````
