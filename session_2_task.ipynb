### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
# 1. Take input in actual numbers (e.g., 1200000 for 12 Lakhs)
ctc = float(input("Enter your annual CTC (e.g., 1200000): "))

# 2. Only deduct PF (3%). HRA and DA stay in your pocket!
salary_after_pf = ctc - (ctc * 0.03)

# 3. Calculate tax based on the remaining salary
if ctc > 2000000:
    taxable_salary = salary_after_pf - (salary_after_pf * 0.3)
elif ctc > 1000000:
    taxable_salary = salary_after_pf - (salary_after_pf * 0.2)
elif ctc > 500000:
    taxable_salary = salary_after_pf - (salary_after_pf * 0.1)
else:
    taxable_salary = salary_after_pf

# 4. Divide by 12 to get the monthly in-hand breakdown
monthly_salary = taxable_salary / 12

print("In Hand monthly salary will be: ₹", round(monthly_salary, 2))
````

### `Problem 5` - Exercise 12: Display Fibonacci series up to 10 terms.

*Note: The Fibonacci Sequence is a series of numbers. The next number is found by adding up the two numbers before it. The first two numbers are 0 and 1. For example, 0, 1, 1, 2, 3, 5, 8, 13, 21. The next number in this series above is 13+21 = 34*

### `Problem 6` - Find the factorial of a given number.

Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: `!`) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```bash
5! = 5 × 4 × 3 × 2 × 1 = 120
```

Output:

```bash
120
```

### `Problem 7` - Reverse a given integer number.

Example:

`Input:`

```bash
76542
```

`Output:`

```bash
24567
```

````python
# Write code here
num = int(input("Enter a number: "))
rev = 0
while num > 0:
  rev = rev * 10 + num % 10
  num = num // 10
print(rev)
````

### `Problem 8`: Take a user input as integer N. Find out the sum from 1 to N. If any number if divisible by 5, then skip that number. And if the sum is greater than 300, don't need to calculate the sum further more. Print the final result. And don't use for loop to solve this problem.

**Example 1:**

`Input:`

```bash
30
```

`Output:`

```bash
276
```

````python
# # Write code here
# num = int(input("Enter a number: "))
# sum = 0
# for i in range(1, num + 1):
#   if i % 5 == 0:
#     continue
#   sum = sum + i
#   if sum > 300:
#     break
# print(sum)

# Take user input as an integer
n = int(input("Enter an integer N: "))

total_sum = 0
current_number = 1

# Run the loop as long as we haven't reached N AND the sum is 300 or less
while current_number <= n and total_sum <= 300:

    # Check if the current number is divisible by 5
    if current_number % 5 == 0:
        current_number += 1
        continue  # Skip the rest of the loop and move to the next number

    # Add the number to our running total
    total_sum += current_number

    # If this addition pushed the sum over 300, break out immediately
    if total_sum > 300:
        break

    # Move to the next number
    current_number += 1

print("Final Sum:", total_sum)
````

### `Problem 9`: Write a program that keeps on accepting a number from the user until the user enters Zero. Display the sum and average of all the numbers.

````python
# Write code here
sum = 0
counter = 0

while True:
  num = int(input("Enter a number: "))
  if num == 0:
    break
  else:
    sum += num
    counter += 1


print("sum = ", sum)
print("average = ", sum / counter)
````

###`Problem 10`: Write a program, which will find all such numbers between 1000 and 3000 (both included) such that each digit of the number is an even number. The numbers obtained should be printed in a space-separated sequence on a single line.



````python
# Write code here

for i in range(1000, 3001):
  all_digits_even = True
  num = i
  while num > 0:
    digit = num % 10
    if digit % 2 != 0:
      all_digits_even = False
      break
    num = num // 10

  if all_digits_even:
    print(i, end="")
````
