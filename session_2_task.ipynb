### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
# 1. Take input in actual numbers (e.g., 1200000 for 12 Lakhs)
ctc = float(input("Enter your annual CTC (e.g., 1200000): "))

# 2. Only deduct PF (3%). HRA and DA stay in your pocket!
salary_after_pf = ctc - (ctc * 0.03)

# 3. Calculate tax based on the remaining salary
if ctc > 2000000:
    taxable_salary = salary_after_pf - (salary_after_pf * 0.3)
elif ctc > 1000000:
    taxable_salary = salary_after_pf - (salary_after_pf * 0.2)
elif ctc > 500000:
    taxable_salary = salary_after_pf - (salary_after_pf * 0.1)
else:
    taxable_salary = salary_after_pf

# 4. Divide by 12 to get the monthly in-hand breakdown
monthly_salary = taxable_salary / 12

print("In Hand monthly salary will be: ₹", round(monthly_salary, 2))
````

###`Problem 11`: A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps.
The trace of robot movement is shown as the following:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
> The numbers after the direction are steps.

> `!` means robot stop there.

**Please write a program to compute the distance from current position after a sequence of movement and original point.**

*If the distance is a float, then just print the nearest integer.*

Example:

`Input`:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
`Output`:
```
2
```

````python
# Write code here
import math

# Initialize coordinates at the origin point (0,0)
x, y = 0, 0

print("Enter the movements (enter '!' to stop):")

while True:
    user_input = input()

    # Check for the stopping condition
    if user_input == "!":
        break

    # Split the input into direction and steps (e.g., "UP 5" -> ["UP", "5"])
    parts = user_input.split()

    # Skip empty inputs if the user hits enter by mistake
    if not parts:
        continue

    direction = parts[0].upper()
    steps = int(parts[1])

    # Update coordinates based on direction
    if direction == "UP":
        y += steps
    elif direction == "DOWN":
        y -= steps
    elif direction == "LEFT":
        x -= steps
    elif direction == "RIGHT":
        x += steps

# Calculate Euclidean distance from the origin (0,0)
distance = math.sqrt(x**2 + y**2)

# Print the nearest integer
print("Distance from origin is: " , round(distance))
````

###`Problem 13`:Print all the Armstrong numbers in a given range.
Range will be provided by the user<br>
Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

````python
# 1. Take range limits from the user
lower_limit = int(input("Enter the lower limit of the range: "))
upper_limit = int(input("Enter the upper limit of the range: "))

print("Armstrong numbers:")

# 2. Loop through every number in the range
for num in range(lower_limit, upper_limit + 1):

    # We need a temporary copy of the number because
    # we will be breaking it down to 0 inside the while loop
    temp = num
    digit_sum = 0

    # 3. Extract each digit one by one using a while loop
    while temp > 0:
        # Get the last digit (e.g., 153 % 10 gives 3)
        digit = temp % 10

        # Cube the digit and add it to our running sum
        digit_sum = digit_sum + (digit * digit * digit)

        # Remove the last digit from the number (e.g., 153 // 10 gives 15)
        temp = temp // 10

    # 4. Special check for 0 (since the while loop won't run if num is 0)
    if num == 0:
        digit_sum = 0

    # 5. If the calculated sum matches the original number, print it
    if digit_sum == num:
        print(num, end=" ")

# Print a clean new line at the very end
print()
````

###`Problem 14`:Calculate the angle between the hour hand and minute hand.

Note: There can be two angles between hands; we need to print a minimum of two. Also, we need to print the floor of the final result angle. For example, if the final angle is 10.61, we need to print 10.

Input:<br>
H = 9 , M = 0<br>
Output:<br>
90<br>
Explanation:<br>
The minimum angle between hour and minute
hand when the time is 9 is 90 degress.

````python
# Write code here
# Take inputs for Hours and Minutes
h = int(input("Enter Hours (H): "))
m = int(input("Enter Minutes (M): "))

# 1. Calculate the position of the hour hand in degrees
# Each hour is 30 degrees, and each minute adds 0.5 degrees
hour_angle = (h * 30) + (m * 0.5)

# 2. Calculate the position of the minute hand in degrees
# Each minute is 6 degrees
minute_angle = m * 6

# 3. Find the absolute difference between the two angles
angle = abs(hour_angle - minute_angle)

# 4. Find the smaller (minimum) of the two possible angles
# (e.g., if the angle is 270 degrees, the reverse side is 360 - 270 = 90)
if angle > 180:
    angle = 360 - angle

# 5. Take the floor value (discard decimal points without rounding up)
final_angle = int(angle)

print("Output:")
print(final_angle)
````

###`Problem 15`:Given two rectangles, find if the given two rectangles overlap or not. A rectangle is denoted by providing the x and y coordinates of two points: the left top corner and the right bottom corner of the rectangle. Two rectangles sharing a side are considered overlapping. (L1 and R1 are the extreme points of the first rectangle and L2 and R2 are the extreme points of the second rectangle).

Note: It may be assumed that the rectangles are parallel to the coordinate axis.

<img src='https://www.geeksforgeeks.org/wp-content/uploads/rectanglesOverlap.png' width='300' height='200'>

````python
# Write code here
def do_rectangles_overlap(l1, r1, l2, r2):
    # Condition 1: Check if one rectangle is completely to the left of the other
    if l1[0] > r2[0] or l2[0] > r1[0]:
        return False

    # Condition 2: Check if one rectangle is completely above the other
    if r1[1] > l2[1] or r2[1] > l1[1]:
        return False

    # If neither condition is true, they must be overlapping or touching
    return True

# --- Example Testing ---

# Rectangle 1: Top-Left (0, 10), Bottom-Right (10, 0)
l1 = [0, 10]
r1 = [10, 0]

# Rectangle 2: Top-Left (5, 5), Bottom-Right (15, -3)
l2 = [5, 5]
r2 = [15, -3]

if do_rectangles_overlap(l1, r1, l2, r2):
    print("The rectangles overlap.")
else:
    print("The rectangles do not overlap.")
````
