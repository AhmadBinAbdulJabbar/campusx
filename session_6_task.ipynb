### **`Problem-1:`** Write a Python function that takes a list and returns a new list with unique elements of the first list.

**Exercise 1:**

Input:

```bash
[1,2,3,3,3,3,4,5]
```

Output:

```bash
[1, 2, 3, 4, 5]
```

````python
def unique_list(lst):
  """
    this function takes a list and returns a new list with unique elements of the first list
  """
  unique_lst = []
  for i in lst:
    if i not in unique_lst:
      unique_lst.append(i)
  return unique_lst
  # return list(set(lst))


print(unique_list([1,2,3,3,3,3,4,5]))
````

### **`Problem-2:`** Write a Python function that accepts a hyphen-separated sequence of words as parameter and returns the words in a hyphen-separated sequence after sorting them alphabetically.

**Example 1:**

Input:
```bash
green-red-yellow-black-white
```

Output:
```bash
black-green-red-white-yellow
```

````python
# Write code here
def sort_words(words):
  words_lst = words.split("-")
  words_lst = sorted(words_lst)
  return "-".join(words_lst)

print(sort_words("green-red-yellow-black-white"))
````

### **`Problem 3:`** Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters.

```
Sample String : 'CampusX is an Online Mentorship Program fOr EnginEering studentS.'
Expected Output :
No. of Upper case characters :  9
No. of Lower case Characters :  47
```

````python
# Write code here
def count_upper_lower(string):
  upper_count = 0
  lower_count = 0
  for i in string:
    if i.isupper():
      upper_count += 1
    elif i.islower():
      lower_count += 1
  return upper_count, lower_count

upper, lower = count_upper_lower('CampusX is an Online Mentorship Program fOr EnginEering studentS.')
print("No. of Upper case characters : ", upper)
print("No. of Lower case Characters : ", lower)
````

### **`Problem 5:`** Write a Python function to check whether a number is perfect or not.

A Perfect number is a number that is half the sum of all of its positive divisors (including itself).

Example :

```
The first perfect number is 6, because 1, 2, and 3 are its proper positive divisors, and 1 + 2 + 3 = 6.
Equivalently, the number 6 is equal to half the sum of all its positive divisors: ( 1 + 2 + 3 + 6 ) / 2 = 6.

The next perfect number is 28 = 1 + 2 + 4 + 7 + 14. This is followed by the perfect numbers 496 and 8128.
```


````python
# Write code here
def is_perfect(num):
  dividers = []
  for i in range(1, num):
    if num % i == 0:
      dividers.append(i)
  if sum(dividers) == num:
    return True
  return False

print(is_perfect(6))
print(is_perfect(28))
print(is_perfect(496))
print(is_perfect(8128))
````

`Problem-7` Write a python function that accepts a string as input and returns the word with most occurence.

```
Input:
hello how are you i am fine thank you
```

```
Output
you -> 2
```

````python
# Write code here
def most_occurence(string):
  words = string.split()
  words_count = {}
  for word in words:
    if word in words_count:
      words_count[word] += 1
    else:
      words_count[word] = 1
  return max(words_count, key=words_count)

print(most_occurence("hello how are you i am fine thank you"))
````

`Problem-8` Write a python function that receives a list of integers and prints out a histogram of bin size 10

```
Input:
[13,42,15,37,22,39,41,50]
```

```
Output:
{11-20:2,21-30:1,31-40:2,41-50:3}
```

````python
# Write code here

def histogram(lst):
  histogram = {}
  for i in range(10, 51, 10):
    count = 0
    for j in lst:
      if j >= i and j < i + 10:
        count += 1
    histogram[f"{i}-{i+10}"] = count
  return histogram

print(histogram([13,42,15,37,22,39,41,50]))


````

`Problem-9` Write a python function that accepts a list of 2D co-ordinates and a query point, and then finds the the co-ordinate which is closest in terms of distance from the query point.

```
List of Coordinates
[(1,1),(2,2),(3,3),(4,4)]
Query Point
(0,0)
```

```
Output
Nearest to (0,0) is (1,1)
```

````python
# Write code here
def distance(point1, point2):
  return ((point1[0] - point2[0]) ** 2 + (point1[1] - point2[1]) ** 2) ** 0.5

def closest_point(lst, query):
  distance_lst = []
  for point in lst:
    distance_lst.append(distance(point, query))
    # return min(distance_lst)
  return lst[distance_lst.index(min(distance_lst))]

print(closest_point([(1,1),(2,2),(3,3),(4,4)], (0,0)))
````

`Problem-10`:Write a python program that receives a list of strings and performs bag of word operation on those strings

https://en.wikipedia.org/wiki/Bag-of-words_model

````python
# Write code here
def bag_of_words(lst):
  words = []
  for string in lst:
    words += string.split()
  words_count = {}
  for word in words:
    if word in words_count:
      words_count[word] += 1
    else:
      words_count[word] = 1
  return words_count

bag_of_words(["hello how are you", "i am fine thank you"])
````

`Problem-14`: Use reduce to convert a 2D list to 1D

````python
import functools

list_2d = [1,2,3,4,[5,6]]

def flatten_list_with_reduce(lst):
    return functools.reduce(lambda acc, x: acc + (x if isinstance(x, list) else [x]), lst, [])

print(flatten_list_with_reduce(list_2d))
````
